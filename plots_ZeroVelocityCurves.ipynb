```python
def plot_pot_3d(X, Y, Z):
    # Plot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap=cm.jet, linewidth=0, antialiased=True)

    # Add contour at base
    ax.contour(X, Y, Z, zdir='z', offset=Z.min(), cmap=cm.jet, levels=200)

    ax.set_xlabel('X (Km)')
    ax.set_ylabel('Y (Km)')
    ax.set_zlabel('U')
    ax.set_title('Gravitational Potential Surface')
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Potential')

    ax.view_init(elev=30, azim=135)

    plt.tight_layout()
    plt.show()


def plot_zer_velocity(X, Y, Z):
    # Plot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111)

    # Add contour at base
    ax.contour(X, Y, Z, zdir='z', offset=Z.min(), cmap=cm.jet, levels=300)

    ax.set_xlabel('X (Km)')
    ax.set_ylabel('Y (Km)')
    ax.set_title('pseudo_potential')
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Potential')

    plt.tight_layout()
    plt.show()
```

# XY 

```python
df = df_data[np.abs(df_data['z_km']) == 0]
# Sort and reshape
x_vals = np.sort(df['x_km'].unique())
y_vals = np.sort(df['y_km'].unique())
X, Y = np.meshgrid(x_vals, y_vals)
Z = df['potential'].values.reshape(len(y_vals), len(x_vals))  # Shape must match
plot_pot_3d(X, Y, Z)
```
